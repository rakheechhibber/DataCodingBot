# outlier detection and removal

```python
def removeoutlier(col):
    sorted(col)
    q1,q3=col.quantile([0.25,0.75])
    IQR=q3-q1
    lowerrange=q1-(1.5 * IQR)
    upperrange = q3 + (1.5 * IQR)
    return lowerrange, upperrange
```

```python
lscore, hscore = removeoutlier(happinessdata['Family'])
print(lscore,hscore)
happinessdata['Family']=np.where(happinessdata['Family']>hscore,hscore,happinessdata['Family'])
happinessdata['Family']=np.where(happinessdata['Family']<lscore,lscore,happinessdata['Family'])
```

```python
lscore, hscore = removeoutlier(happinessdata['Generosity'])
print(lscore,hscore)
happinessdata['Generosity']=np.where(happinessdata['Generosity']>hscore,hscore,happinessdata['Generosity'])
happinessdata['Generosity']=np.where(happinessdata['Generosity']<lscore,lscore,happinessdata['Generosity'])
```
